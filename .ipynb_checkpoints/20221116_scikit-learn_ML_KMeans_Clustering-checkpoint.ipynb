# scikit-learn Machine Learning K-Means Clustering <br> 사이킷런 머신러닝 K-평균 군집화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Collection

```python
file_url = 'https://raw.githubusercontent.com/dev-EthanJ/scikit-learn_Machine_Learning/main/data/example_cluster.csv'
df = pd.read_csv(file_url)
```

```python
df.head()
```

```python
sns.scatterplot(x='var_1', y='var_2', data=df)

plt.show()
```

## Training Model

```python
from sklearn.cluster import KMeans

kmeans_model = KMeans(n_clusters=3, random_state=814)
kmeans_model.fit(df)

pred = kmeans_model.predict(df)
pred
```

```python
df['label'] = pred
```

```python
sns.scatterplot(x='var_1', y='var_2', data=df, hue='label', palette='rainbow')

plt.show()
```

- inertia (이너셔)

    - 각 그룹에서의 중심과 데이터간의 거리에 대한 합

```python
kmeans_model.inertia_
```

```python
new_kmeans_model = KMeans(n_clusters=500, random_state=814)
new_kmeans_model.fit(df)
new_kmeans_model.inertia_
```

```python
df['new_label'] = new_kmeans_model.predict(df)

df['new_label']
```

```python
sns.scatterplot(x='var_1', y='var_2', data=df,
                hue='new_label', palette='rainbow')

plt.show()
```

## Data Collection

```python
file_url = 'https://raw.githubusercontent.com/dev-EthanJ/scikit-learn_Machine_Learning/main/data/customer.csv'
df = pd.read_csv(file_url)

df.head()
```

```python
df.describe(include='all')
```

```python
df.cc_num.nunique()
```

## Data pre-processing

- 고객들을 군집화 → 분류별 소비금액 (100명)

```python
customer_dummies = pd.get_dummies(df, columns=['category'])

customer_dummies.info()
```

```python
cate_list = customer_dummies.columns[2:]

customer_dummies.head()
```

```python
for col in cate_list:
    customer_dummies[col] = customer_dummies[col] * customer_dummies['amt']

customer_dummies
```

```python
customer_agg = customer_dummies.groupby('cc_num').sum()

customer_agg.head()
```

```python
customer_agg.info()
```

## Training Model

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled = scaler.fit_transform(customer_agg)
scaled
```

```python
scaled_df = pd.DataFrame(scaled, columns=customer_agg.columns,
                         index=customer_agg.index)

scaled_df
```

```python
distance = list()

for k in range(2, 10):
    k_model = KMeans(n_clusters=k)
    k_model.fit(scaled_df)
    labels = k_model.predict(scaled_df)
    distance.append(k_model.inertia_)
```

```python
sns.lineplot(x=range(2, 10), y=distance)

plt.show()
```

## Evaluating Model

- `inertia`: 그룹의 중심으로부터 외부 요소간의 거리

- `silhouette_score`: cluster 내부의 평균 거리 + 최근접 다른 cluster data의 평균 거리

```python
from sklearn.metrics import silhouette_score
```

```python
def k_model_score(k):
    k_model = KMeans(n_clusters=k)
    k_model.fit(scaled_df)
    labels = k_model.predict(scaled_df)
    return silhouette_score(scaled_df, labels)
```

```python
silhouette = list()

for k in range(2, 10):
    silhouette.append(k_model_score(k))

silhouette
```

- 실루엣 계수: 더 높은 값일수록 더 좋은 classifier

```python
sns.lineplot(x=range(2, 10), y=silhouette)

plt.show()
```

```python
k_model = KMeans(n_clusters=4)
k_model.fit(scaled_df)
labels = k_model.predict(scaled_df)

scaled_df['label'] = labels

scaled_df.head()
```

```python
scaled_df_mean = scaled_df.groupby('label').mean()

scaled_df_mean.head()
```

```python
scaled_df_count = scaled_df.groupby('label').count()['amt']

scaled_df_count
```

```python
scaled_df_count = scaled_df_count.rename('count')

scaled_df_count
```

```python
scale_df = scaled_df_mean.join(scaled_df_count)

scale_df
```

### 2.3. í›ˆë ¨ì…‹ `train set`, ì‹œí—˜ì…‹ `test set`

**ë°ì´í„° ë¶„í•  ì‘ì—…**
1. ì¢…ì†ë³€ìˆ˜ì™€ ë…ë¦½ë³€ìˆ˜ ë¶„ë¦¬
2. í•™ìŠµì…‹(Train set)ê³¼ ì‹œí—˜ì…‹(Test set) ë¶„ë¦¬

||ë…ë¦½ë³€ìˆ˜|ì¢…ì†ë³€ìˆ˜|
|-|-|-|
|í•™ìŠµì…‹|X_train|y_train|
|ì‹œí—˜ì…‹|X_test|y_test|

#### ë…ë¦½ë³€ìˆ˜ì™€ ì¢…ì†ë³€ìˆ˜
* **ë…ë¦½ë³€ìˆ˜**
  * ì˜ˆì¸¡ì— ì‚¬ìš©ë˜ëŠ” ì¬ë£Œì™€ ê°™ì€ ë³€ìˆ˜ë“¤
  * í”¼ì³ ë³€ìˆ˜(Feature variable), í”¼ì³(Feature)
* **ì¢…ì†ë³€ìˆ˜**
  * ì˜ˆì¸¡ì„ í•˜ë ¤ëŠ” ëŒ€ìƒ ë³€ìˆ˜
  * ëª©í‘œ ë³€ìˆ˜ (Target variable), íƒ€ê¹ƒ

_Why?_ : ì§€ë„ í•™ìŠµì— ì†í•˜ëŠ” ëª¨ë¸ì€ ë…ë¦½ë³€ìˆ˜ë¥¼ ì˜ˆì¸¡í•˜ëŠ” ê²ƒì´ë¯€ë¡œ, ëª¨ë¸ë§í•  ë•Œ ì–´ë–¤ ë³€ìˆ˜ê°€ ì¢…ì†ë³€ìˆ˜ì¸ì§€ ëª…í™•íˆ ì•Œë ¤ì£¼ì–´ì•¼ í•¨ â†’ ë…ë¦½ë³€ìˆ˜ì™€ ì¢…ì†ë³€ìˆ˜ë¥¼ ê°ê° ë³„ë„ì˜ ë°ì´í„°ë¡œ ì…ë ¥ ë°›ìŒ

- `train set`ê³¼ `test set`ì„ ë‚˜ëˆ„ëŠ” ì´ìœ :

  - `train set`ìœ¼ë¡œ í›ˆë ¨ ëœ ê²°ê³¼ë¡œ ë§Œë“¤ì–´ì§„ ëª¨ë¸ì„ ê°€ì§€ê³  ì‹œí—˜ì„ í•˜ë©´ `train set`ì— ëŒ€í•´ì„œëŠ” ë§ì¶œ ìˆ˜ ë°–ì— ì—†ìŒ

  - í•™ìŠµì…‹ê³¼ ì‹œí—˜ì…‹ì„ êµ¬ë¶„í•˜ì§€ ì•Šê³  ì˜ˆì¸¡ ëª¨ë¸ì„ ë§Œë“¤ ê²½ìš° ìƒˆë¡œìš´ ë°ì´í„°ì— ëŒ€í•œ ì˜ˆì¸¡ë ¥ì„ ê²€ì¦í•  ìˆ˜ ì—†ìŒ

    - ì „ì²´ ë°ì´í„°ë¥¼ ê°€ì§€ê³  ëª¨ë¸ë§(í•™ìŠµ)ì„ í•˜ê³ , ë˜ ë‹¤ì‹œ ì „ì²´ ë°ì´í„°ì— ëŒ€í•´ì„œ ì˜ˆì¸¡ê°’ì„ ë§Œë“¤ì–´ì„œ ì¢…ì†ë³€ìˆ˜ì™€ ë¹„êµí–ˆê¸° ë•Œë¬¸ì— íŠ¹ì • ë°ì´í„°ì— ëŒ€í•´ ê³¼ìµœì  ë˜ì—ˆì„ ìˆ˜ ìˆìŒ

    - ì‹œí—˜ì…‹ì„ í†µí•´ ê²€ì¦í•˜ì§€ ì•Šì€ ìƒíƒœë¼ëŠ” ë¶ˆí™•ì‹¤ì„±ì„ ì¤„ì¼ ìˆ˜ ìˆìŒ
    
      - Dataì˜ ì¼ë¶€ë¥¼ `test set`ìœ¼ë¡œ ë…ë¦½ì‹œì¼œì„œ, `train set`ê³¼ `test set`ì„ ë‚˜ëˆˆë‹¤.

      - ì¼ë°˜ì ìœ¼ë¡œ **train set:test set**ì„ ê°ê° 7:3, 8:2 ì •ë„ë¡œ ë‚˜ëˆ” (ì „ì²´ ë°ì´í„°ê°€ ì ì„ ê²½ìš° í•™ìŠµì…‹ì„ ìš°ì„ ì‹œ)

````python
df.columns
````

````python
# X: ë…ë¦½ë³€ìˆ˜, y: ì¢…ì†ë³€ìˆ˜
X = df.drop('expenses', axis=1)
y = df.expenses
````

- scikit-learn import

````python
from sklearn.model_selection import train_test_split
````

````python
help(train_test_split)
````

````python
# test_size: ë¹„ìœ¨ > test set(ì‹œí—˜ì…‹)ì˜ ratio
# random_state: seed value > ì„ì˜ë¡œ ê²°ì •ë˜ëŠ” ê°’ì„ íŠ¹ì • ê°’ìœ¼ë¡œ ìœ ì§€ë˜ê²Œ í•˜ëŠ” ê°’

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2, random_state = 200
)

len(train_test_split(X, y, test_size = 0.2, random_state=200))
````

  - ë…ë¦½ë³€ìˆ˜ì™€ ì¢…ì†ë³€ìˆ˜ë¥¼ *fit()* í•¨ìˆ˜ì— ì¸ìˆ˜ë¡œ ì£¼ì–´ í•™ìŠµ

|ì•Œê³ ë¦¬ì¦˜ ì„ íƒ| |ëª¨ë¸ ìƒì„±| |ëª¨ë¸ í•™ìŠµ| |ëª¨ë¸ì„ ì‚¬ìš©í•œ ì˜ˆì¸¡|

|ë‹¤ì–‘í•œ ë¨¸ì‹ ëŸ¬ë‹<br>ì•Œê³ ë¦¬ì¦˜ ì¤‘ ì„ íƒ í›„<br>ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸|â†’|ëª¨ë¸ë§ì— ì‚¬ìš©í• <br>ëª¨ë¸ ìƒì„±|â†’|*fit()* í•¨ìˆ˜ì— í•™ìŠµì…‹ì˜<br>ë…ë¦½ë³€ìˆ˜ì™€ ì¢…ì†ë³€ìˆ˜ë¥¼<br>ì¸ìˆ˜ë¡œ ëŒ€ì…|â†’|train_test_split()<br>í•¨ìˆ˜ì— í‰ê°€ì…‹ì˜ <br>ë…ë¦½ë³€ìˆ˜ë¥¼ ì¸ìˆ˜ë¡œ ëŒ€ì…

````python
from sklearn.linear_model import LinearRegression
````

````python
model = LinearRegression()
````

### 3.1. `model.fit(X_train, y_train)`

- ëª¨ë¸ì„ í•™ìŠµì‹œí‚¤ëŠ” í•¨ìˆ˜ : Dataë¥¼ Modelì•ˆì— ë„£ì–´ì„œ ë…ë¦½ë³€ìˆ˜`X_train`ê³¼ ì¢…ì†ë³€ìˆ˜`y_train`ê°„ì˜ ê´€ê³„ë“¤ ë¶„ì„,   
ìƒˆë¡œìš´ Dataë¥¼ ì˜ˆì¸¡í•  ìˆ˜ ìˆë„ë¡ í•™ìŠµì‹œí‚´

````python
model.fit(X_train, y_train)
````

### 3.2. `model.predict(X_test)`

- ëª¨ë¸ì„ í™œìš©í•´ì„œ ì˜ˆì¸¡

````python
pred = model.predict(X_test)

pred
````

### 4.3. RMSE & RÂ²(ê²°ì •ê³„ìˆ˜)

|í‰ê°€ì§€í‘œ|ì„¤ëª…|
|-|-|
|MAE (Mean Absolute Error, í‰ê·  ì ˆëŒ€ ì˜¤ì°¨)|- ì‹¤ì ¯ê°’ê³¼ ì˜ˆì¸¡ê°’ ì‚¬ì´ì˜ ì˜¤ì°¨ì— ì ˆëŒ“ê°’ì„ ì”Œìš´ ë’¤ ì´ì— ëŒ€í•œ í‰ê· ì„ ê³„ì‚°<br>- 0ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ì¢‹ìŒ|
|MSE (Mean Squared Error, í‰ê·  ì œê³± ì˜¤ì°¨)|- ì‹¤ì ¯ê°’ê³¼ ì˜ˆì¸¡ê°’ ì‚¬ì´ì˜ ì˜¤ì°¨ë¥¼  ì œê³±í•œ ë’¤ ì´ì— ëŒ€í•œ í‰ê· ì„ ê³„ì‚°<br>- 0ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ì¢‹ìŒ|
|RMSE (Root Mean Absolute Error, ë£¨íŠ¸ í‰ê·  ì œê³± ì˜¤ì°¨)<br>|- MSEì— ë£¨íŠ¸ë¥¼ ì”Œìš´ ê°’<br>- 0ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ì¢‹ìŒ<br>- ì—°ì†í˜• ë³€ìˆ˜ë¥¼ ì˜ˆì¸¡í•  ë•Œ ê°€ì¥ ì¼ë°˜ì ìœ¼ë¡œ ì‚¬ìš©ë˜ëŠ” í‰ê°€ì§€í‘œ|
|RÂ²|- ê²°ì •ê³„ìˆ˜<br>- ë…ë¦½ë³€ìˆ˜ê°€ ì¢…ì†ë³€ìˆ˜ë¥¼ ì–¼ë§ˆë§Œí¼ ì„¤ëª…í•´ ì£¼ëŠ”ì§€ ë‚˜íƒ€ë‚´ëŠ” ì§€í‘œ, ì¦‰ ì„¤ëª…ë ¥<br>- 1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ì¢‹ìŒ|

- `mean_squared_error(y_test, pred)`

````python
from sklearn.metrics import mean_squared_error
````

````python
# MSE(í‰ê· ì œê³±ì˜¤ì°¨): ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ ì‚¬ì´ì˜ ì˜¤ì°¨ì˜ ì œê³±ì˜ í‰ê· 
mean_squared_error(y_test, pred)
````

````python
# RMSE(í‰ê· ì œê³±ê·¼ì˜¤ì°¨): Root MSE
# mean_squared_error(y_test, X_pred) ** 0.5
mean_squared_error(y_test, pred, squared=False)
````

- `model.score(X_train, y_train)`

- $R^2 = \frac{SSR}{SST}$

  - **SST** (Sum of Squares Total) : ì´ë³€ë™

  - **SSR** (Sum of Squares Regression) : ì„¤ëª… ê°€ëŠ¥í•œ ë³€ë™

    - ëª¨ë¸ì´ ì–¼ë§ˆë‚˜ ê°’ì˜ ë³€ë™ì„ ì˜ ì„¤ëª…í•˜ëŠ”ê°€? = ì˜ˆì¸¡ì„ ì˜í•˜ëŠ”ê°€?

````python
# R^2 = ê²°ì •ê³„ìˆ˜: ëª¨ë¸ì´ ì–¼ë§ˆë‚˜ ì¢…ì†ë³€ìˆ˜-ë…ë¦½ë³€ìˆ˜ì˜ ë³€ë™ì„ ì˜ ì„¤ëª…í•˜ëŠ”ê°€(ì˜ˆì¸¡ ì •í™•ë„)ë¥¼ ìˆ˜ì¹˜í™”í•œ ê°’
R_squared = model.score(X_train, y_train)

R_squared
````
